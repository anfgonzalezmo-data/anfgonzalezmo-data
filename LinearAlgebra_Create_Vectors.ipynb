## DOT PRODUCT CALCULATION

```python
## Create two vectors
v1 = np.array([1,2,5,7,0])
v2 = np.array([-3,2,0,-1,6])
```

```python
# Method 1 Element wise multiplication between the elements of the vector and then sum
dp1 = sum(np.multiply(v1,v2))
dp1
```

```python
# Method 2 using the numpy function
dp2 = np.dot(v1,v2)
dp2
```

```python
# Method 3, using matrix multiplication
dp3 = np.matmul(v1,v2)
dp3
```

```python
# Method 4, Using a loop over the vectors
dp4 = 0
for i in range(0,len(v2)):
  # multiply the corresponding elements and sum
  dp4 = dp4 + v1[i]*v2[i]

print(dp4)
```

## DOT PRODUCT PROPERTIES ASSOCIATIVE AND DISTRIBUTIVE

```python
# Create random vectors
n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

print(a,b,c)
```

```python
# Calculate the dot product to prove the DISTRIBUTIVE
res1 = np.dot(a, (b+c))
res2 = np.dot(a, c) + np.dot(a,b)

print(res1, res2)
```

```python
# Calculate the dot product to prove is NOT ASSOCIATIVE
res1 = np.dot(np.dot(a,b),c)
res2 = np.dot(a,np.dot(b,c))

print([res1, res2])
```

DOT PRODUCT FROM GEOMETRIC PERSPECTIVE

```python
# Create two vectors
v1 = np.array([2,4,-3])
v2 = np.array([0,-3,-3])
```

```python
# Calculate the angle in radians between the vectors
ang = np.arccos(np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2)))
ang
```

```python
# Plot the vectors
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], label='v1')
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], label='v2')

# Set the axis limits
ax.set_xlim([-6, 6])
ax.set_ylim([-6, 6])
ax.set_zlim([-6, 6])

# Set the title and legend
plt.title('Test dot product vectors')
ax.legend()

# Display the plot
plt.show()
```

```python
# equivalence of algebraic and geometric dot product
```

```python
# Algebraic
dp_a = np.dot(v1,v2)
# Geometric
dp_g = np.linalg.norm(v1)*np.linalg.norm(v2)*np.cos(ang)
```

```python
print(dp_a, dp_g)
```

# Prove the CAUCHY_SCHWARZ Inequality
# $|a^{t}b|\leq \left \| a \right \|\left \| b \right \|$

```python
# create three vectors, and 
a = np.random.randint(0,10,4)
b = np.random.randint(0,10,4)
c = np.random.randint(1,4,1)*a
print(a,b,c)
# Compute the dot products
atb = np.dot(a,b)
atc = np.dot(a,c)
abnorm = np.linalg.norm(a)*np.linalg.norm(b)
acnorm = np.linalg.norm(a)*np.linalg.norm(c)

# demonstrate the inequality
print(f'{np.abs(atb)}, {abnorm:.2f}')

# demonstrate the equality
print(f'{np.abs(atc)}, {acnorm:.2f}')
```

# PROF IF THE DOT PRODUCT **SIGN** IS CHANGED BY SCALAR MULTIPLICATION

```python
# 1. Create vectors
r1 = np.random.randint(-10,10,3)
r2 = np.random.randint(-10,10,3)

# 2. assign the scalars
s1 = 2
s2 = s1 * -1

# 3.Compute the dot product between the vectors
dot1 = np.dot(r1,r2)
# 4.Compute the dot product between the scaled vectors
dot2 = np.dot(s1*r1,s1*r2)

# 5.Compute the dot product between the scaled vectors with changed sign
dot3 = np.dot(r1*s2,r2*s2)

print(f'Vector 1: {r1} \nVector 2: {r2}\nScalar 1: {s1}\nScalar2: {s2}\n')
print(f'Dot product: {dot1} \nDot product scaled: {dot2}\nDot product scaled sign changed scalar: {dot3}')
```

## THE OUTER PRODUCT OF A VECTOR

```python
# Define the vectors
o1 = np.array(np.random.randint(-10,10,4))
o2 = np.array(np.random.randint(-10,10,3))

print(o1, o2)

# compute with the numpy function
print(np.outer(o1, o2))

# compute with a for cicle
op = np.zeros((len(o1), len(o2)))
for i in range(0, len(o1)):
  for j in range(0, len(o2)):
    op[i,j] = o1[i]*o2[j]

print(op)
```
